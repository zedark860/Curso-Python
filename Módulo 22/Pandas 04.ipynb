- Vamos criar então agora uma lista de Clientes

```python
clientes = pd.read_csv('Contoso - Clientes.csv', sep=';')

clientes.info()

lista_clientes = []

for _, cliente in clientes.iterrows():
    # devolve duas iterações no loop, nunca esquecer de colocar duas variáveis para receber os dados
    primeiro_nome = cliente['Primeiro Nome']
    segundo_nome = cliente['Sobrenome']
    lista_clientes.append(primeiro_nome + ' ' + segundo_nome)

print(lista_clientes)
```

- Vamos criar agora uma lista com os produtos e as quantidades de vendas dele, caso a gente queira analisar só os produtos (independente de data ou de cliente)
