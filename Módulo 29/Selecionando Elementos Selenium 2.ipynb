```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os

# Procure por id, classe única e por último xpath
# Xpath não é a melhor forma, pois pode mudar de uma hora pra outra

navegador.get("https://www.hashtagtreinamentos.com/cursos-hashtag-treinamentos")
```

#### Selecionar pela tag

- Vamos selecionar o t�tulo

```python
from selenium.webdriver.common.by import By

titulo = navegador.find_element(By.TAG_NAME, 'h1').text
print(titulo)
```

#### Selecionar pelo Partial Link Text (ou LINK_TEXT)
- Quero conseguir pegar o n�mero de whatsapp de contato

```python
numero_whatsapp = navegador.find_element(By.PARTIAL_LINK_TEXT, 'WhatsApp').text
print(numero_whatsapp)
```

#### Selecionar pelo name
- Preencher o formul�rio

```python
navegador.find_element(By.NAME, 'email').send_keys("pythonimpressionador@gmail.com")
```
