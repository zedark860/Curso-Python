```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\formulario.html"
navegador.get(arquivo)
```

#### Bot�o Padr�o (clicar em bot�o)

```python
navegador.find_element(By.XPATH, '/html/body/form/input[1]').click()
```

```python
alerta = navegador.switch_to.alert
alerta.accept()
```

#### Bot�o de Sele��o estilo Checkbox (clicar no bot�o)

```python
# Clica no botão
navegador.find_element(By.XPATH, '/html/body/form/input[2]').click()
```

```python
# Verifica valor do botão
valor = navegador.find_element(By.XPATH, '/html/body/form/input[2]').is_selected()
print(valor)
```

#### Bot�o de Sele��o de Cores (enviar valor)

```python
# Preencher a cor
# utilize sempre o hexadecimal
navegador.find_element(By.XPATH, '/html/body/form/input[4]').send_keys("#2143E8")
```

```python
# Verificar qual cor foi selecionada

valor = navegador.find_element(By.XPATH, '/html/body/form/input[4]').get_attribute("value")
print(valor)
```
