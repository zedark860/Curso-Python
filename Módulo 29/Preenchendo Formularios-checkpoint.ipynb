```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\formulario.html"
navegador.get(arquivo)
```

```python
import time

# primeiro abrindo a lista
navegador.find_element(By.XPATH, '/html/body/form/select[1]').click()

time.sleep(1)

# depois clicando no elemento
navegador.find_element(By.XPATH, '/html/body/form/select[1]/option[1]').click()
```

```python
# com select
elemento = navegador.find_element(By.TAG_NAME, 'select')
elemento_select = Select(elemento)
```

```python
#.select_by_index(index)
# seleciona pelo indice

#.select_by_visible_text("text")
# seleciona pelo texto visível na lista, pode ser qualquer um

#.select_by_value(value)
# seleciona pelo valor da lista, pode ser qualquer um

#.deselect_all()
# quando se pode escolher várias opções na lista suspensa
```

```python
elemento_select.select_by_index(2)

# ou

elemento_select.select_by_value('c')

# ou

elemento_select.select_by_visible_text('C')
```

```python
# para saber se consegue selecionar vários elementos
print(elemento_select.is_multiple)
```

```python
# ler item selecionado

# é o primeiro item que ele ver selecionado
item = elemento_select.first_selected_option

# mostra o valor
print(item.get_attribute('value'))

# ou

item = navegador.find_element(By.XPATH, '/html/body/form/select[1]/option[1]').get_attribute('value')
```

```python
# lista de itens
lista_itens = elemento_select.all_selected_options
print(lista_itens[0].text)
```
